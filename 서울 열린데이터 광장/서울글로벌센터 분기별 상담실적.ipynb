```python
q_list = ['19년 1분기 실적','19년 2분기 실적','19년 3분기 실적','19년 4분기 실적',
        '20년 1분기 실적','20년 2분기 실적','20년 3분기 실적','20년 4분기 실적',
        '21년 1분기 실적','21년 2분기 실적','21년 3분기 실적','21년 4분기 실적',
        '22년 1분기 실적','22년 2분기 실적','22년 3분기 실적','22년 4분기 실적']

type1 = ['출입국','노무','보건/복지','행정','교육','기타']
type2 = ['주거','안전','금융','여성/가족','실 근무일','비즈니스','정보통신','교통','문화/관광']
colors1 = sns.color_palette('hls',6)
colors2 = sns.color_palette('Set1',9)

def convert_shape(df,c):
    new_df = pd.DataFrame([c for _ in range(16)],columns=['기관/유형'])
    new_df['분기'] = q_list
    new_df['건수'] = df[df['기관/유형']==c].sort_values('년')[['1/4분기 실적','2/4분기 실적','3/4분기 실적','4/4분기 실적']].values.reshape(16)
    return new_df

convert_df1 = pd.DataFrame()
for c in type1:
    convert_df1 = pd.concat([convert_df1,convert_shape(df,c)])
convert_df1.reset_index(drop=True,inplace=True)

convert_df2 = pd.DataFrame()
for c in type2:
    convert_df2 = pd.concat([convert_df2,convert_shape(df,c)])
convert_df2.reset_index(drop=True,inplace=True)


fig,axes = plt.subplots(2,1,figsize=(24,16))
g1=sns.barplot(data=convert_df1,x='분기',y='건수',hue='기관/유형',ax=axes[0],palette=colors1)
plt.title(f'분기별 총 건수')
for container in g1.containers:
    g1.bar_label(container)

g2=sns.barplot(data=convert_df2,x='분기',y='건수',hue='기관/유형',ax=axes[1],palette=colors2)
plt.title(f'분기별 총 건수')
for container in g2.containers:
    g2.bar_label(container)
```
