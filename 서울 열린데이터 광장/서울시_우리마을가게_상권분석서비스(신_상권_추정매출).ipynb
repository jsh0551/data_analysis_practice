## 분석 결과
**각 상권의 연도/분기 매출액,매출건수,점포수는?**

- 연도가 지남에 따라 골목상권,관광특구,발달상권의 매출액과 이용건수는 감소하지만 전통시장은 꾸준히 증가하는 추세다.

- 매출액과 이용건수는 2020년을 제외하고 대체로 3,4분기가 좀 더 높은 경향이 있다.

- 점포수는 연도나 분기에 따른 두드러지는 특징이 나타나지 않앗다.

**각 상권의 시간대별 매출액과 매출건수는?**

- 매출액과 이용건수는 11시-21시 사이가 70%-80%를 차지한다.

- 골목상권,관광특구,발달상권은 매출액이 17시-21시가 가장 높고 전통시장은 14시-17시가 가장 높다.

- 이용 건수의 경우 모든 상권에서 17시-21시 시간대가 높게 나왔다.

**각 상권의 연령대별 매출액과 매출건수는?**

- 매출액은 40대,30대,50대 순으로 높았고 이용건수는 20대,30대,40대 순으로 높았다.

- 골목상권과 발달상권은 30대-50대가, 관광특구는 20대-40대가, 전통시장은 40대-60대가 매출금액 비율이 높았다.

- 이용건수는 골목상권,관광특구,발달상권에서 20대,30대,40대 순으로 높았고 전통시장은 20대,50대,60대 순으로 높았다.

**각 상권의 성별 매출액과 매출건수는?**

- 골목상권과 발달상권은 남성이, 관광특구와 전통시장은 여성이 매출액과 이용건수가 높았다.

**가장 많이 찾는 서비스업종과 상권 top10과 bottom10?**

- 명동, 강남역, 가산디지털단지, 노량진, 홍대입구 순으로 이용 건수가 높았고 자양종합시장(기능상실), 지하쇼핑센터 등이 이용 건수가 낮았다.

- 편의점, 한식음식점, 커피음료, 슈퍼마켓이 이용건수가 높았고 건축물청소, 회계사사무소, 한복점, 중고차판매 직종이 이용 건수가 낮았다.

## 분셕 결과에 대한 생각

- 2020년 하반기가 다른 년도와 다른 매출 양상을 보인 것은 코로나 유행의 영향으로 보인다.

- 다른 상권과 달리 전통 시장의 매출이 꾸준히 증가하고 있는데, 이용 건수가 20대가 가장 높다는 점을 보았을 때 유튜브 등 여러 미디어 매체의 영향을 받은 것으로 보인다.

- 전통 시장의 주 매출원이 50대-60대이고, 가장 매출이 높은 시간대가 14시-17시라는 점을 보았을 때 전통 시장은 중장년층 주부 고객들이 주로 장을 보는 영역이라는 생각이 들었다.

- 상권마다 주고객층의 시간, 연령, 성별이 상이하다는 점을 활용한다면 각 상권에 속한 자영업자들의 매출을 향상시키기 위한 전략을 효과적으로 설계할 수 있을 것 같다.
