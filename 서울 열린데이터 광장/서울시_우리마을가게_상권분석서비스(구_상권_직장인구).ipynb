```python
sub_df = df.groupby(by=['기준_년월_코드'],as_index=False)['총_직장_인구_수'].sum()
sub_df1 = df.groupby(by=['기준_년월_코드'],as_index=False)['남성_직장_인구_수'].sum()
sub_df2 = df.groupby(by=['기준_년월_코드'],as_index=False)['여성_직장_인구_수'].sum()
fig,axes = plt.subplots(3,1,figsize=(12,15))
sns.barplot(data=sub_df, x='기준_년월_코드',y='총_직장_인구_수',ax=axes[0]).set(title='연도별 총 직장 인구 수')
sns.barplot(data=sub_df1, x='기준_년월_코드',y='남성_직장_인구_수',ax=axes[1]).set(title='연도별 남성 직장 인구 수')
sns.barplot(data=sub_df2, x='기준_년월_코드',y='여성_직장_인구_수',ax=axes[2]).set(title='연도별 여성 직장 인구 수')
plt.tight_layout()
```

```python
age_cols = ['연령대_10_직장_인구_수', '연령대_20_직장_인구_수', '연령대_30_직장_인구_수', '연령대_40_직장_인구_수', '연령대_50_직장_인구_수', '연령대_60_이상_직장_인구_수']

fig,axes = plt.subplots(8,1,figsize=(12,40))
for i in range(8):
    axes[i].bar(age_cols,df[df['기준_년월_코드']==2014+i][age_cols].sum().values)
    axes[i].set_title(f'{i+2014}년 연령대별 직장 인구 수')
    axes[i].set_ylim([0,4e6])
plt.tight_layout()
```

```python
m_cols = ['남성연령대_10_직장_인구_수', '남성연령대_20_직장_인구_수','남성연령대_30_직장_인구_수', '남성연령대_40_직장_인구_수', '남성연령대_50_직장_인구_수','남성연령대_60_이상_직장_인구_수']
w_cols = ['여성연령대_10_직장_인구_수', '여성연령대_20_직장_인구_수','여성연령대_30_직장_인구_수', '여성연령대_40_직장_인구_수', '여성연령대_50_직장_인구_수','여성연령대_60_이상_직장_인구_수']
fig,axes = plt.subplots(2,2,figsize=(24,10))
axes[0][0].bar(m_cols,df[df['기준_년월_코드']==2014][m_cols].sum().values)
axes[0][0].set_title(f'2014년 남성 연령대별 직장 인구 수')
axes[0][0].set_ylim([0,1.8e6])
axes[0][1].bar(w_cols,df[df['기준_년월_코드']==2014][w_cols].sum().values,color='r')
axes[0][1].set_title(f'2014년 여성 연령대별 직장 인구 수')
axes[0][1].set_ylim([0,1.8e6])
axes[1][0].bar(m_cols,df[df['기준_년월_코드']==2021][m_cols].sum().values)
axes[1][0].set_title(f'2021년 남성 연령대별 직장 인구 수')
axes[1][0].set_ylim([0,1.8e6])
axes[1][1].bar(w_cols,df[df['기준_년월_코드']==2021][w_cols].sum().values,color='r')
axes[1][1].set_title(f'2021년 여성 연령대별 직장 인구 수')
axes[1][1].set_ylim([0,1.8e6])
plt.tight_layout()
```

```python
fig,axes = plt.subplots(1,2,figsize=(24,6))
sub_df1 = df.groupby(by=['상권_구분_코드_명'],as_index=False)['남성_직장_인구_수'].sum()
sub_df2 = df.groupby(by=['상권_구분_코드_명'],as_index=False)['여성_직장_인구_수'].sum()
sns.barplot(data=sub_df1,x='상권_구분_코드_명',y='남성_직장_인구_수',ax=axes[0]).set(title='상권별 남성 직장 인구 수')
sns.barplot(data=sub_df2,x='상권_구분_코드_명',y='여성_직장_인구_수',ax=axes[1]).set(title='상권별 여성 직장 인구 수')
```

```python
area = df['상권_구분_코드_명'].unique()
age_cols = [ '연령대_10_직장_인구_수', '연령대_20_직장_인구_수','연령대_30_직장_인구_수', '연령대_40_직장_인구_수', '연령대_50_직장_인구_수','연령대_60_이상_직장_인구_수']
fig,axes = plt.subplots(4,1,figsize=(16,20))
axes[0].bar(age_cols,df[df['상권_구분_코드_명']==area[0]][age_cols].sum().values)
axes[0].set_title(f'{area[0]} 연령대별 직장 인구 수')
axes[1].bar(age_cols,df[df['상권_구분_코드_명']==area[1]][age_cols].sum().values)
axes[1].set_title(f'{area[1]} 연령대별 직장 인구 수')
axes[2].bar(age_cols,df[df['상권_구분_코드_명']==area[2]][age_cols].sum().values)
axes[2].set_title(f'{area[2]} 연령대별 직장 인구 수')
axes[3].bar(age_cols,df[df['상권_구분_코드_명']==area[3]][age_cols].sum().values)
axes[3].set_title(f'{area[3]} 연령대별 직장 인구 수')
```

## 분석 결과

**각 연도의 총 직장 인구 수는?**

- 2014년에 가장 낮고, 2018년까지 상승하다 2019년부터 감소하는 추세다.

**각 연도의 성별/연령대별 인구 수는?**

- 시간이 지나면서 40대-60대의 비율이 높아졌다.

**각 성별 연령대 분포는?**

- 남성은 전체 인구 중 30대-40대가 가장 높은 비율을 차지하고, 여성은 20대-30대가 가장 높은 비율을 차지한다.

**각 상권의 성별 직장 인구 수는?**

- 두 성별 다 발달상권, 골목상권, 관광특구, 전통시장 순으로 많다.

**각 상권의 연령대별 직장 인구 수는?**

- 각 상권 모두 30대-40대 비율이 가장 높지만, 골목상권과 전통시장은 발달상권과 관광특구에 비해 40대-60대 비율이 높다.

## 분석 결과에 대한 생각

- 전체 직장 인구 수의 감소, 40대-60대 직장인 비율 증가 등을 고려했을 때 고령화와 저출산으로 인한 영향이 드러나고 있는 것 같다.

- 가장 경제 활동이 활발한 30대를 제외하면, 연령대별로 선호하는 상권이 어느정도 구분되는 것 같다.
