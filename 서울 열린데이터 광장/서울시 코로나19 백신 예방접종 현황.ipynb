## 전처리

```python
from datetime import datetime,timedelta

def get_year(x):
    return x.year

def get_month(x):
    return x.month

def get_day(x):
    return x.day

def get_week(x):
    date = x
    first_date = date.replace(day=1)
    start = first_date.weekday()
    while first_date.weekday() != start:
        first_date += timedelta(days=1)
    return ((date-first_date).days+start)//7 + 1

def get_weekday(x):
    weekday = x.weekday()
    if weekday%7 == 0:
        return '월'
    elif weekday%7 == 1:
        return '화'
    elif weekday%7 == 2:
        return '수'
    elif weekday%7 == 3:
        return '목'
    elif weekday%7 == 4:
        return '금'
    elif weekday%7 == 5:
        return '토'
    else:
        return '일'
```

```python
f_date = datetime.strptime(df['접종일'].loc[0],'%Y.%m.%d')
s_date = datetime.strptime(df['접종일'].loc[len(df)-1],'%Y.%m.%d')

test_df = pd.DataFrame({'접종일':[f_date]})
while f_date>s_date:
    f_date -= timedelta(days=1)
    tmp = pd.DataFrame({'접종일':[f_date]})
    test_df = pd.concat([test_df,tmp],ignore_index=True)
test_df.head()
```

```python
df['접종일'] = pd.to_datetime(df['접종일'])
```

```python
df = pd.merge(test_df,df, how = 'left', on = '접종일')
df.head()
```

```python
date_df = pd.DataFrame()
date_df['연도'] = df['접종일'].apply(get_year)
date_df['월'] = df['접종일'].apply(get_month)
date_df['일'] = df['접종일'].apply(get_day)
date_df['주차'] = df['접종일'].apply(get_week)
date_df['요일'] = df['접종일'].apply(get_weekday)
date_df.head()
```

```python
ffill_cols = ['접종대상자','1차접종 누계','1차접종률(%)','2차접종 누계','2차접종률(%)','3차접종 누계','3차접종률(%)','4차접종 누계','4차접종률(%)']
df[ffill_cols] = df[ffill_cols].fillna(method='ffill')
```

```python
def combine_2rd_columns(col_1, col_2,col_3=None):
    result = str(int(col_1))
    if col_3:
        result += '.' + str(int(col_2)) + ' ' + str(int(col_3)) + '주차'
        return result
    else:
        result += '.' + str(int(col_2))
        return result
```

```python
fig,axes = plt.subplots(4,1,figsize=(20,20))
axes[0].set_ylim([0,2000])
axes[1].set_ylim([0,2000])
axes[2].set_ylim([0,2000])
axes[3].set_ylim([0,2000])

df1 = df.groupby(by = ['연도','월'],as_index=False)['당일 1차접종자 수'].sum()
df2 = df.groupby(by = ['연도','월'],as_index=False)['당일 2차접종자 수'].sum()
df3 = df.groupby(by = ['연도','월'],as_index=False)['당일 3차접종자 수'].sum()
df4 = df.groupby(by = ['연도','월'],as_index=False)['당일 4차접종자 수'].sum()
df1['당일 1차접종자 수'] = np.sqrt((df1['당일 1차접종자 수']))
df2['당일 2차접종자 수'] = np.sqrt((df2['당일 2차접종자 수']))
df3['당일 3차접종자 수'] = np.sqrt((df3['당일 3차접종자 수']))
df4['당일 4차접종자 수'] = np.sqrt((df4['당일 4차접종자 수']))
df1['날짜'] = df1.apply(lambda x: combine_2rd_columns(x['연도'], x['월']), axis=1)
df2['날짜'] = df2.apply(lambda x: combine_2rd_columns(x['연도'], x['월']), axis=1)
df3['날짜'] = df3.apply(lambda x: combine_2rd_columns(x['연도'], x['월']), axis=1)
df4['날짜'] = df4.apply(lambda x: combine_2rd_columns(x['연도'], x['월']), axis=1)
sns.lineplot(data=df1,x='날짜',y='당일 1차접종자 수',ax=axes[0]).set(title='월별 1차 접종자 수(square scale)')
sns.lineplot(data=df2,x='날짜',y='당일 2차접종자 수',ax=axes[1]).set(title='월별 2차 접종자 수(square scale)')
sns.lineplot(data=df3,x='날짜',y='당일 3차접종자 수',ax=axes[2]).set(title='월별 3차 접종자 수(square scale)')
sns.lineplot(data=df4,x='날짜',y='당일 4차접종자 수',ax=axes[3]).set(title='월별 4차 접종자 수(square scale)')

plt.tight_layout()
```

```python
q_df = df.groupby(by = ['연도','월'], as_index=False)[['1차접종 누계','2차접종 누계','3차접종 누계','4차접종 누계']].max()
q_df['날짜'] = q_df.apply(lambda x: combine_2rd_columns(x['연도'], x['월']), axis=1)
q_df
```

```python
fig,axes = plt.subplots(1,1,figsize=(22,12))
axes.set_ylim([0,2.6e7])
bars = q_df.iloc[:,[2,3,4,5]].plot(kind='bar', stacked=True,ax = axes)
xticks = plt.xticks([i for i in range(len(q_df))],labels=q_df['날짜'].values)
for i in range(len(q_df)):
    height = axes.patches[i].get_height() + axes.patches[i+21].get_height() + axes.patches[i+42].get_height() + axes.patches[i+63].get_height()
    h1 = round(axes.patches[i].get_height()/height*100,2)
    h2 = round(axes.patches[i+21].get_height()/height*100,2)
    h3 = round(axes.patches[i+42].get_height()/height*100,2)
    h4 = round(axes.patches[i+63].get_height()/height*100,2)
    height = int(height)
    ann_height = height*1.04
    if i == 0:
        ann_height = height*2
    elif i < 4:
        ann_height = height*1.4
    axes.annotate(str(height),(axes.patches[i].get_x()*0.99, ann_height))
    axes.annotate(str(h1)+'%',(axes.patches[i].get_x()*1, 0.5*h1/100*height*1.02))
    axes.annotate(str(h2)+'%',(axes.patches[i].get_x()*1, (h1+0.5*h2)/100*height*1.02))
    if h3>0.01:
        axes.annotate(str(h3)+'%',(axes.patches[i].get_x()*1, (h1+h2+0.5*h3)/100*height*1.02))
    if h4>0.01:
        axes.annotate(str(h4)+'%',(axes.patches[i].get_x()*1, (h1+h2+h3)/100*height*1.02))
plt.legend(loc='upper left')
```
