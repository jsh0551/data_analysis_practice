## 전처리

```python
df.drop(['사무실주소','사업금액(억원)'],axis=1, inplace=True)
idx1 = df[df['프로젝트 주소'].isnull()].index
df.drop(idx1,inplace=True)
idx2 = df[df['위치좌표(위도)'].isnull()].index
df.drop(idx2,inplace=True)
idx3 = df[df['위치좌표(경도)'].isnull()].index
df.drop(idx3,inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
def fill_area(c1,c2):
    if type(c1) == float or type(c2) == str:
        return c2
    area = c1.split()
    if len(area)==1:
        return c2
    else:
        if area[0][-1] == '구':
            return area[0]
        elif area[1][-1]!='시' or area[1][-1]!='군':
            return '서울지역외'
        else:
            return c2



df['자치구 구분'] = df.apply(lambda x : fill_area(x['프로젝트 주소'],x['자치구 구분']), axis=1)
```

```python
idx4 = df[df['자치구 구분'].isnull()].index
df.drop(idx4,inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df.isnull().sum()
```

```python
def to_start_date(x):
    date = x.split('~')
    start = datetime.strptime(date[0],'%Y-%m-%d')
    m = str(100+start.month)[1:]
    return str(start.year) + '.' + m

def to_end_date(x):
    date = x.split('~')
    end = datetime.strptime(date[1],'%Y-%m-%d')
    m = str(100+end.month)[1:]
    return str(end.year) + '.' + m

def to_time(x):
    date = x.split('~')
    return (datetime.strptime(date[1],'%Y-%m-%d') - datetime.strptime(date[0],'%Y-%m-%d'))


df['사업착수일'] = df['사업기간'].apply(to_start_date)
df['사업종료일'] = df['사업기간'].apply(to_end_date)
df['사업기간(개월)'] = (df['사업기간'].apply(to_time)).dt.days//30
```

```python
df.drop(['사업착수일(계약일)','사업기간'],axis=1,inplace=True)
```

**사업 기간 분포는? 프로젝트 종료 여부 별 사업 기간은?**

- 전체의 20프로 이상이 한달 이내에 사업이 종료되고, 사업 기간 대부분은 12개월 안에 끝난다.

**프로젝트 종료 여부 별 사업 착수일/종료일?**

- 착수일 기준으로 봤을 때 대부분의 사업은 2022년에 분포되어 있다.

- 진행중 사업은 착수일이 2022년 전체적으로 고르게 분포되어 있다.

- 종료된 사업은 착수일이 2022년 9월-11월에 집중되어 있다.

**자치구 분포는?**

- 성동구가 가장 사업수를 제외하면 사업 수가 꾸준히 있는 지역과 거의 없는 지역이 눈에 띄게 구분된다.

**프로젝트 종료 여부 별 자치구 분포는??**

- 종료 여부에 따른 분포의 차이는 거의 없다.

**서울외지역과 서울지역 분포는?**

- 사업의 61프로가 서울지역, 39프로가 서울외지역이다.

**서올외지역 제외 위치 분포는?**

**사업 많은/적은 자치구 탑5?**

- 성동구가 500건 이상으로 가장 많고 마포구, 중구, 양천구, 서대문구 순서로 많다.

- 금천구와 서초구강남구가 1건으로 가장 적고 강북구, 광진구, 강동구 순으로 적다.

## 분석 결과에 대한 추측

- 개발 정도, 건물 낙후 정도 등 공사 건수가 적은 지역과 많은 지역이 극명하게 나뉘는 여러 이유가 존재할 것이다.

- 특히 성동구가 공사 건수가 많이 분포되어 있어 성동구와 다른 자치구의 특징을 잘 파악한다면 공사가 필요한 지역을 예측할 수 있을 것이다.
